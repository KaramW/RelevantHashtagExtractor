```python
import pandas as pd
import tensorflow_hub as hub
import numpy as np
```

```python
# Step 1: Read the file
hashtags_df = pd.read_csv('/content/drive/MyDrive/Clients/Maven/DataAnalysis/data_source/Hashtags/hashtags_without#_marged_data.csv')
```

```python
# Step 2: Preprocess the textual content
caption_content= "Unique Right-to-Left Script: Arabic breaks the mold with its right-to-left script, a trait that makes it stand out in the world of languages. Context is Key: Arabic writing often omits short vowels. So, understanding Arabic means mastering the art of context. Language of Depth: With a vocabulary exceeding 12.3 million words, Arabic's richness towers above English's modest 600,000 words. Quran's Linguistic Legacy: Classical Arabic, the language of the Quran, has left a significant imprint on modern Arabic. Diverse Dialects: Arabic's vibrant tapestry includes over 30 distinct dialects. Each Arabic-speaking nation adds its unique linguistic touch. A World Language: Arabic, the official language in 26 countries, ranks as the 5th most spoken language globally. Arabic's Gift to Spanish: Spanish vocabulary is imbued with over 4000 Arabic words, a testament to Arabic's wide-reaching influence. Root and Branch: Arabic's root system is a fascinating linguistic feature. The majority of words spring from a three-letter root, subtly altering in meaning with added letters. True Arabic Numerals: Arabic numerals in the West are Indian in origin. In Arabic-speaking lands, a different numeral system applies. The Art of Arabic Calligraphy: This traditional art form lends Arabic an additional layer of aesthetic charm. Feeling Inspired? This is just the start! There's a vast universe of Arabic language and culture awaiting your exploration. Maven Training & Consulting offers beginner-friendly and advanced courses to make your Arabic learning experience both fun and rewarding. Ready to dive in? Kickstart your journey today! Connect with us to discover more about our language courses and how they can assist you in exploring the captivating world of Arabic"
```

```python
# Step 3: Calculate semantic similarity
embed = hub.load("https://tfhub.dev/google/universal-sentence-encoder/4")
text_embedding = embeddings = embed([caption_content])
relevant_hashtags = []
for hashtag in hashtags_df['Hashtags']:
    hashtag_embedding = embed([hashtag])[0]
    similarity_score = np.inner(text_embedding, hashtag_embedding)
    if similarity_score > 0.10:  # Set a threshold for relevance
        relevant_hashtags.append(hashtag)
```

```python
len(relevant_hashtags)
```

```python
# Step 5: Select top relevant hashtags
top_hashtags = sorted(
    relevant_hashtags,
    key=lambda x: hashtags_df.loc[hashtags_df['Hashtags'].isin([x]), 'Search Volume (Average)'].iloc[0],
    reverse=True
)[:10]
```
